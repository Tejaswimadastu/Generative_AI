### Task 2: Fixed Size Chunking

```python
def fixed_size_chunking(text, chunk_size):
    chunks = []
    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i + chunk_size])
    return chunks

chunk_sizes = [100, 200, 300]

for size in chunk_sizes:
    print(f"\n--- Fixed Size Chunking (Chunk Size = {size}) ---")
    chunks = fixed_size_chunking(text, size)
    for i, chunk in enumerate(chunks):
        print(f"Chunk Number: {i+1}")
        print(f"Chunk Content: {chunk}")
        print(f"Chunk Length: {len(chunk)}")
        print("---")


print("\n--- Analysis of Fixed Size Chunking ---")
print("Are sentences being split? (Yes/No): Yes, often.")
print("Is context preserved? (Yes/No): Partially, context can be lost at chunk boundaries.")
print("What are the limitations?: ")
print("1. Arbitrary splits: Sentences and ideas can be cut mid-way.")
print("2. Context loss: Important information might be divided across chunks, making retrieval harder.")
print("3. Suboptimal for varying content density: A fixed size may be too small for dense sections and too large for sparse sections.")
```

### Task 3: Recursive Chunking

```python
import sys

# Install LangChain if not already installed
if 'google.colab' in sys.modules:
    %pip install langchain-text-splitters

from langchain_text_splitters import RecursiveCharacterTextSplitter

def recursive_chunking(text, chunk_size, chunk_overlap):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        is_separator_regex=False,
    )
    chunks = text_splitter.create_documents([text])
    return chunks

# Experiment with different chunk sizes and overlaps
recursive_params = [
    {"chunk_size": 200, "chunk_overlap": 20},
    {"chunk_size": 200, "chunk_overlap": 50},
    {"chunk_size": 300, "chunk_overlap": 50},
]

for params in recursive_params:
    size = params["chunk_size"]
    overlap = params["chunk_overlap"]
    print(f"\n--- Recursive Chunking (Chunk Size = {size}, Chunk Overlap = {overlap}) ---")
    chunks = recursive_chunking(text, size, overlap)
    for i, chunk in enumerate(chunks):
        print(f"Chunk Number: {i+1}")
        print(f"Chunk Content: {chunk.page_content}")
        print("---")

print("\n--- Analysis of Recursive Chunking --- ")
print("How overlap improves context retention: ")
print("Overlap helps in preserving context by ensuring that sentences or phrases that fall on chunk boundaries are included in both adjacent chunks. This reduces the risk of losing important information or the semantic meaning of a sentence that might be split in a fixed-size chunking approach. When retrieving information, if a query spans across a chunk boundary, the overlap increases the likelihood that relevant parts of the context are present in at least one of the retrieved chunks, thus improving the overall retrieval quality.")
```

### Task 4: Sentence-Based Chunking

```python
import sys

# Install NLTK if not already installed
if 'google.colab' in sys.modules:
    %pip install nltk

import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Download punkt_tab resource as suggested by the error

def sentence_chunking(text):
    sentences = nltk.sent_tokenize(text)
    return sentences

print("--- Sentence-Based Chunking ---")
sentences = sentence_chunking(text)
for i, sentence in enumerate(sentences):
    print(f"Sentence Number: {i+1}")
    print(f"Sentence Content: {sentence}")
    print("---")

print("\n--- Analysis of Sentence-Based Chunking ---")
print("Does sentence chunking preserve meaning better than fixed-size chunking? (Yes/No): Yes.")
print("Explanation: Sentence chunking naturally preserves the complete meaning of each individual sentence, as it avoids splitting a sentence mid-way. Fixed-size chunking, on the other hand, can arbitrarily cut sentences, leading to fragmented information and potential loss of context within a chunk. By keeping sentences intact, sentence chunking ensures that each chunk represents a coherent unit of thought, which generally leads to better context preservation and improved retrieval quality for queries that align with sentence structures.")
```

### Task 5: Semantic Chunking

```python
import re

def semantic_chunking(text):
    # Define patterns for each policy section
    leave_policy_pattern = r"Leave Policy\n.*?\n(?=Work From Home Policy|Travel Policy|Medical Insurance Policy|$)"
    wfh_policy_pattern = r"Work From Home Policy\n.*?\n(?=Travel Policy|Medical Insurance Policy|$)"
    travel_policy_pattern = r"Travel Policy\n.*?\n(?=Medical Insurance Policy|$)"
    medical_policy_pattern = r"Medical Insurance Policy\n.*?\n(?=Leave Policy|Work From Home Policy|Travel Policy|$)"

    chunks = []

    # Find leave policy
    match = re.search(leave_policy_pattern, text, re.DOTALL)
    if match: # Use if 'match' is not None
        chunks.append({"name": "Leave Policy Section", "content": match.group(0).strip()})
        text = text[match.end():] # Remove the matched part from the text

    # Find work from home policy
    match = re.search(wfh_policy_pattern, text, re.DOTALL)
    if match:
        chunks.append({"name": "Work From Home Section", "content": match.group(0).strip()})
        text = text[match.end():]

    # Find travel policy
    match = re.search(travel_policy_pattern, text, re.DOTALL)
    if match:
        chunks.append({"name": "Travel Policy Section", "content": match.group(0).strip()})
        text = text[match.end():]

    # Find medical insurance policy (and any remaining text)
    match = re.search(medical_policy_pattern, text, re.DOTALL)
    if match:
        chunks.append({"name": "Medical Insurance Section", "content": match.group(0).strip()})

    return chunks

print("--- Semantic Chunking ---")
semantic_chunks = semantic_chunking(text)
for i, chunk_data in enumerate(semantic_chunks):
    print(f"Semantic Chunk Number: {i+1}")
    print(f"Semantic Chunk Name: {chunk_data['name']}")
    print(f"Semantic Chunk Content: {chunk_data['content']}")
    print("---")

print("\n--- Analysis of Semantic Chunking ---")
print("Why semantic chunking may improve retrieval quality: ")
print("Semantic chunking groups content based on its inherent meaning or topic, ensuring that all related information stays together in a single chunk. This is highly beneficial for retrieval systems because when a user's query relates to a specific topic (e.g., 'leave policy'), the entire, relevant section of information is retrieved at once, rather than fragmented pieces across multiple chunks. This leads to more comprehensive and accurate answers, reducing the need for the retrieval system to piece together context from disparate sources. It aligns well with how humans understand and categorize information, making the retrieval process more intuitive and effective.")
```

### Task 7: Retrieval Simulation

```python
import sys

# Install necessary libraries
if 'google.colab' in sys.modules:
    %pip install -q transformers sentence-transformers
    %pip install -q faiss-cpu

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import faiss # For efficient similarity search

# Load a pre-trained sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Define queries
queries = [
    "How many casual leaves are provided?",
    "Can employees work from home?",
    "What is the travel reimbursement process?",
    "Who is covered under medical insurance?"
]

# Function to retrieve the most relevant chunk
def retrieve_most_relevant_chunk(query, chunks, chunk_method_name):
    if not chunks:
        print(f"No chunks available for {chunk_method_name}.")
        return

    # Extract content from Langchain Document objects if applicable
    chunk_contents = [c.page_content if hasattr(c, 'page_content') else c for c in chunks]

    # Generate embeddings for chunks and query
    chunk_embeddings = model.encode(chunk_contents, convert_to_tensor=True)
    query_embedding = model.encode(query, convert_to_tensor=True)

    # Convert embeddings to numpy arrays for FAISS
    chunk_embeddings_np = chunk_embeddings.cpu().numpy()
    query_embedding_np = query_embedding.cpu().numpy()

    # Create a FAISS index
    dimension = chunk_embeddings_np.shape[1]
    index = faiss.IndexFlatL2(dimension) # Using L2 distance for similarity
    index.add(chunk_embeddings_np)

    # Search for the most similar chunk
    D, I = index.search(query_embedding_np.reshape(1, -1), 1) # D is distances, I is indices

    most_relevant_chunk_index = I[0][0]
    most_relevant_chunk = chunk_contents[most_relevant_chunk_index]

    print(f"Query: {query}")
    print(f"Chunking Method: {chunk_method_name}")
    print(f"Retrieved Chunk: {most_relevant_chunk}")
    print("---")

# --- Perform retrieval for each chunking strategy ---

# Fixed Size Chunking (using the text and sizes from Task 2)
fixed_size_chunks_100 = fixed_size_chunking(text, 100)
fixed_size_chunks_200 = fixed_size_chunking(text, 200)
fixed_size_chunks_300 = fixed_size_chunking(text, 300)

print("\n--- Retrieval Simulation: Fixed Size Chunking (100) ---")
for query in queries:
    retrieve_most_relevant_chunk(query, fixed_size_chunks_100, "Fixed Size (100)")

print("\n--- Retrieval Simulation: Fixed Size Chunking (200) ---")
for query in queries:
    retrieve_most_relevant_chunk(query, fixed_size_chunks_200, "Fixed Size (200)")

print("\n--- Retrieval Simulation: Fixed Size Chunking (300) ---")
for query in queries:
    retrieve_most_relevant_chunk(query, fixed_size_chunks_300, "Fixed Size (300)")

# Recursive Chunking (using chunks from Task 3)
recursive_chunks_200_20 = recursive_chunking(text, 200, 20)
recursive_chunks_200_50 = recursive_chunking(text, 200, 50)
recursive_chunks_300_50 = recursive_chunking(text, 300, 50)

print("\n--- Retrieval Simulation: Recursive Chunking (200, 20) ---")
for query in queries:
    retrieve_most_relevant_chunk(query, recursive_chunks_200_20, "Recursive (200, 20)")

print("\n--- Retrieval Simulation: Recursive Chunking (200, 50) ---")
for query in queries:
    retrieve_most_relevant_chunk(query, recursive_chunks_200_50, "Recursive (200, 50)")

print("\n--- Retrieval Simulation: Recursive Chunking (300, 50) ---")
for query in queries:
    retrieve_most_relevant_chunk(query, recursive_chunks_300_50, "Recursive (300, 50)")

# Sentence-Based Chunking (using sentences from Task 4)
print("\n--- Retrieval Simulation: Sentence-Based Chunking ---")
for query in queries:
    retrieve_most_relevant_chunk(query, sentences, "Sentence-Based")

# Semantic Chunking (using semantic_chunks from Task 5)
semantic_chunk_contents = [c['content'] for c in semantic_chunks] # Extract content from dicts
print("\n--- Retrieval Simulation: Semantic Chunking ---")
for query in queries:
    retrieve_most_relevant_chunk(query, semantic_chunk_contents, "Semantic")
```

### Task 8: Recommendation Report

### Recommendation for Chunking Strategies in RAG Systems

Based on the comparative analysis of various chunking strategies (Fixed Size, Recursive, Sentence-Based, and Semantic), here are recommendations for different types of RAG systems:

---

#### 1. HR Policy Assistant

*   **Recommended Strategy:** **Semantic Chunking**

*   **Justification:** HR policies are typically structured into distinct sections (e.g., Leave Policy, Work From Home Policy, Travel Policy, Medical Insurance Policy). Semantic chunking excels at identifying and preserving these natural boundaries, ensuring that an entire policy section, with its complete context, is treated as a single unit. When an employee asks a question about a specific policy (e.g., "How many casual leaves are provided?"), semantic chunking will retrieve the entire 'Leave Policy' section, providing comprehensive and accurate information. This minimizes the risk of fragmented answers or missing context, which is crucial for sensitive HR-related inquiries where precision is paramount. While sentence-based chunking is also good for context, semantic chunking goes a step further by grouping related sentences into meaningful policy blocks.

---

#### 2. Legal Document Assistant

*   **Recommended Strategy:** **Sentence-Based Chunking with potential for Recursive Chunking (small overlap) for definitions/clauses**

*   **Justification:** Legal documents often rely on precise wording, and the meaning can change drastically if sentences are split incorrectly. Sentence-based chunking ensures that each legal statement or clause remains intact, preserving its exact meaning. This is critical for legal accuracy and avoiding misinterpretation. For documents with complex definitions or interconnected clauses, a small degree of recursive chunking (e.g., `chunk_size` = 200, `chunk_overlap` = 20) could be beneficial to slightly bridge context between very closely related sentences, without breaking the integrity of individual sentences. The primary goal is to maintain the integrity of legal statements as distinct, coherent units, which sentence-based chunking achieves effectively.

---

#### 3. Medical Document Assistant

*   **Recommended Strategy:** **Semantic Chunking or Recursive Chunking (moderate overlap)**

*   **Justification:** Medical documents, such as patient records, research papers, or clinical guidelines, often contain distinct sections (e.g., patient history, diagnosis, treatment plan, medication details) or paragraphs detailing specific conditions or procedures. Semantic chunking would be ideal if these sections are clearly delineated. However, if the text is more free-flowing but still organized by topic within paragraphs, recursive chunking with a moderate overlap (e.g., `chunk_size` = 300, `chunk_overlap` = 50) would be highly effective. This approach balances the need to keep related medical information together while allowing some overlap to capture dependencies between adjacent paragraphs or sentences, which is common in medical narratives or explanations. This helps in retrieving comprehensive information about a specific medical aspect without losing crucial surrounding context.

---

#### 4. Research Paper Assistant

*   **Recommended Strategy:** **Recursive Chunking (larger chunk size, moderate overlap)**

*   **Justification:** Research papers typically have longer, more complex paragraphs and sections (e.g., Introduction, Methodology, Results, Discussion). A larger chunk size (e.g., `chunk_size` = 300 or more, depending on typical paragraph length) is appropriate to capture the detailed explanations and arguments within these sections. A moderate overlap (e.g., `chunk_overlap` = 50) is important to ensure continuity between chunks, especially when ideas are developed across multiple sentences or paragraphs. This allows the RAG system to retrieve substantial blocks of information that can provide a complete picture of a specific concept, experiment, or finding, which is essential for understanding academic content. Fixed-size chunking would be too arbitrary, and purely sentence-based or semantic might break up complex arguments that span several sentences or lack clear structural markers for semantic grouping.
